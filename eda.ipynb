# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('SLDC_Data/2017/10/01-10-2017.csv')
```

```python
data.head()
```

```python
data.plot(x='time', y='value');
```

```python
plt.scatter(data['time'], data['value'])
plt.show()
```

```python
data.plot(subplots=True, figsize=(20, 10)); plt.legend(loc='best');
```

```python
data.plot(kind='bar'); plt.axhline(0, color='k');
```
